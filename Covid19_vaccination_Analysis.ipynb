```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("country_vaccinations.csv")
data
```

```python
data.isnull().sum()
```

```python
df1 = data[['country','people_vaccinated','people_fully_vaccinated']]
df1
```

```python
df1_new = df1.fillna(0)
df1_new
```

```python
data_1 = df1_new.groupby('country').sum().reset_index()
data_1
```

```python
data_2 = data_1.nlargest(5,['people_vaccinated','people_fully_vaccinated'])
data_2
```

```python
people_vaccinated_top5 = data_1[['country','people_vaccinated']].nlargest(5,'people_vaccinated')
people_vaccinated_top5
```

```python
people_vaccinatedfully_top5 = data_1[['country','people_fully_vaccinated']].nlargest(5,'people_fully_vaccinated')
people_vaccinatedfully_top5
```

```python
data_2.index
```

```python
%matplotlib inline
```

```python
people_vaccinated_top5['people_vaccinated'] = people_vaccinated_top5['people_vaccinated'].apply(pd.to_numeric)
```

```python
plt.ylabel('Countries')
plt.xlabel('People Vaccinated')
plt.title('Top 5 people Vaccinated Countries')
plt.barh(width = people_vaccinated_top5['people_vaccinated'], y = people_vaccinated_top5['country'], 
              color = 'indigo',edgecolor = 'black',tick_label = people_vaccinated_top5['country'])
```

```python
plt.ylabel('Countries')
plt.xlabel('People Fully Vaccinated')
plt.title('Top 5 fully Vaccinated Countries')
plt.barh(width = people_vaccinatedfully_top5['people_fully_vaccinated'], y = people_vaccinatedfully_top5['country'], 
              color = 'red',edgecolor = 'black',tick_label = people_vaccinatedfully_top5['country'])
```

```python
vaccine_data = data[['country','vaccines']]
vaccine_data
```

```python
d1 = vaccine_data.groupby(['country','vaccines']).count().reset_index()
d1
```

```python
d1[['Vaccine1','Vaccine2','Vaccine3']] = d1['vaccines'].str.split(',',expand = True)
d1
```

```python
d1 = d1.drop(columns = ['vaccines'], axis = 1)
```

```python
d1
```

```python
d2 = d1[['country','Vaccine1']]
d2 = d2.rename(columns = {'Vaccine1':'Vaccines'})
d2
```

```python
d3 = d1[['country','Vaccine2']]
d3 = d3.rename(columns = {'Vaccine2':'Vaccines'})
d3
```

```python
d4 = d1[['country','Vaccine3']]
d4 = d4.rename(columns = {'Vaccine3':'Vaccines'})
d4
```

```python
vaccine_data_new = pd.concat([d2,d3,d4], ignore_index = True)
vaccine_data_new
```

```python
vaccine_data_new = vaccine_data_new.dropna()
vaccine_data_new.loc[:,'Vaccines'] = vaccine_data_new['Vaccines'].str.strip()
vaccine_data_new
```

```python
# Top Vaccines used by Countries for Covid
vaccine_d = vaccine_data_new.groupby(['Vaccines']).count().nlargest(5,'country').reset_index()
vaccine_d
```

```python
vaccine_d['country'] = vaccine_d['country'].apply(pd.to_numeric)
vaccine_d
```

```python
explode = (0.1,0.1,0,0,0)
plt.pie(vaccine_d['country'], explode = explode, labels=vaccine_d['Vaccines'],
        shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
